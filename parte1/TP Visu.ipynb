```python
def calcularedad(ano,mes,dia):
    edadA= 2018-ano
    edadM= 4-mes
    edadD= 8-dia
    if (edadM>0):
        if(edadD>0):
            return edadA
    return edadA-1

P_Edad['edad'] = P_Edad['fechanacimiento'].map(lambda x: calcularedad(x.year,x.month,x.day))
```

```python
P_Edad = P_Edad.loc[P_Edad['edad']>18,:]
P_Edad= P_Edad.loc[P_Edad['edad']<80,:]
```

```python
P_Edad['edad'].value_counts()
```

```python
fig=plt.figure()
ax = fig.add_subplot(111)

plt.title('Edad postulantes: Historigram',fontsize=16,fontweight='bold')
ax.set_xlabel('Edad')
P_Edad['edad'].plot.hist(figsize=[10,10], bins=60)
```
